```python
from IPython.display import display, Javascript, Image, Audio
from google.colab.output import eval_js
from base64 import b64decode
from PIL import Image as PILImage
import pytesseract
from gtts import gTTS
from googletrans import Translator
import os

def take_photo(filename='photo.jpg', quality=0.8):
    js = Javascript('''
    async function takePhoto(quality) {
        const div = document.createElement('div');
        const capture = document.createElement('button');
        capture.textContent = 'Capture';
        div.appendChild(capture);
        const video = document.createElement('video');
        video.style.display = 'block';
        const stream = await navigator.mediaDevices.getUserMedia({video: true});
        document.body.appendChild(div);
        div.appendChild(video);
        video.srcObject = stream;
        await video.play();
        google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);
        await new Promise((resolve) => capture.onclick = resolve);
        const canvas = document.createElement('canvas');
        canvas.width = video.videoWidth;
        canvas.height = video.videoHeight;
        canvas.getContext('2d').drawImage(video, 0, 0);
        stream.getVideoTracks()[0].stop();
        div.remove();
        return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
    display(js)
    data = eval_js(f'takePhoto({quality})')
    binary = b64decode(data.split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename

def extract_text_from_image(image_path):
    with PILImage.open(image_path) as image:
        return pytesseract.image_to_string(image)

def text_to_speech(text, output_file, lang='en'):
    tts = gTTS(text=text, lang=lang)
    tts.save(output_file)
    return output_file

def translate_text(text, dest_lang='hi'):
    translator = Translator()
    return translator.translate(text, dest=dest_lang).text

def main():
    try:
        # Take photo
        filename = take_photo()
        print(f'Saved to {filename}')
        display(Image(filename))

        # Extract text from image
        text = extract_text_from_image(filename)
        print("Extracted text:")
        print(text)

        if not text:
            print("No text was extracted from the image.")
            return

        # Convert text to speech
        audio_file = text_to_speech(text, "/content/output_en.mp3")
        print("English audio file created successfully!")
        display(Audio(audio_file))

        # Translate text to Hindi
        hindi_text = translate_text(text, 'hi')
        print("Translated text (Hindi):")
        print(hindi_text)

        # Convert translated text to speech
        hindi_audio_file = text_to_speech(hindi_text, "/content/output_hi.mp3", 'hi')
        print("Hindi audio file created successfully!")
        display(Audio(hindi_audio_file))

    except Exception as err:
        print(f"An error occurred: {str(err)}")

if __name__ == "__main__":
    main()
```

```python
!sudo apt install tesseract-ocr
!pip install pytesseract
!pip install gtts
!pip install googletrans==3.1.0a0 speech_recognition

!apt-get install -y tesseract-ocr
!pip install pytesseract Pillow gtts googletrans==3.1.0a0 SpeechRecognition

from IPython.display import display, Javascript, Image, Audio
from google.colab.output import eval_js
from base64 import b64decode
from PIL import Image as PILImage
import pytesseract
from gtts import gTTS
from googletrans import Translator
import os
import speech_recognition as sr

def take_photo(filename='photo.jpg', quality=0.8):
    js = Javascript('''
    async function takePhoto(quality) {
        const div = document.createElement('div');
        const capture = document.createElement('button');
        capture.textContent = 'Capture';
        div.appendChild(capture);
        const video = document.createElement('video');
        video.style.display = 'block';
        const stream = await navigator.mediaDevices.getUserMedia({video: true});
        document.body.appendChild(div);
        div.appendChild(video);
        video.srcObject = stream;
        await video.play();
        google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);
        await new Promise((resolve) => capture.onclick = resolve);
        const canvas = document.createElement('canvas');
        canvas.width = video.videoWidth;
        canvas.height = video.videoHeight;
        canvas.getContext('2d').drawImage(video, 0, 0);
        stream.getVideoTracks()[0].stop();
        div.remove();
        return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
    display(js)
    data = eval_js(f'takePhoto({quality})')
    binary = b64decode(data.split(',')[1])
    with open(filename, 'wb') as f:
        f.write(binary)
    return filename

def extract_text_from_image(image_path):
    with PILImage.open(image_path) as image:
        return pytesseract.image_to_string(image)

def text_to_speech(text, output_file, lang='en'):
    tts = gTTS(text=text, lang=lang)
    tts.save(output_file)
    return output_file

def translate_text(text, dest_lang='hi'):
    translator = Translator()
    return translator.translate(text, dest=dest_lang).text

def speech_to_text():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening for speech...")
        audio = recognizer.listen(source)
        try:
            text = recognizer.recognize_google(audio)
            print("You said: " + text)
            return text
        except sr.UnknownValueError:
            print("Google Speech Recognition could not understand audio")
        except sr.RequestError as e:
            print("Could not request results from Google Speech Recognition service; {0}".format(e))
    return ""

def main():
    try:
        # Option to take photo or use speech
        choice = input("Do you want to [1] take a photo or [2] speak? (1/2): ")

        if choice == '1':
            # Take photo
            filename = take_photo()
            print(f'Saved to {filename}')
            display(Image(filename))

            # Extract text from image
            text = extract_text_from_image(filename)
            print("Extracted text:")
            print(text)

            if not text:
                print("No text was extracted from the image.")
                return

        elif choice == '2':
            # Capture speech and convert to text
            text = speech_to_text()
            if not text:
                return

        else:
            print("Invalid choice. Exiting.")
            return

        # Convert text to speech
        audio_file = text_to_speech(text, "/content/output_en.mp3")
        print("English audio file created successfully!")
        display(Audio(audio_file))

        # Translate text to Hindi
        hindi_text = translate_text(text, 'hi')
        print("Translated text (Hindi):")
        print(hindi_text)

        # Convert translated text to speech
        hindi_audio_file = text_to_speech(hindi_text, "/content/output_hi.mp3", 'hi')
        print("Hindi audio file created successfully!")
        display(Audio(hindi_audio_file))

    except Exception as err:
        print(f"An error occurred: {str(err)}")

if __name__ == "__main__":
    main()
```

```python
!sudo apt install tesseract-ocr
!pip install pytesseract gtts googletrans==3.1.0a0 speech_recognition gradio

!apt-get install -y tesseract-ocr
!pip install pytesseract Pillow gtts googletrans==3.1.0a0 SpeechRecognition

from PIL import Image as PILImage
import pytesseract
from gtts import gTTS
from googletrans import Translator
import os
import speech_recognition as sr
import gradio as gr

def extract_text_from_image(image):
    return pytesseract.image_to_string(image)

def text_to_speech(text, lang='en'):
    tts = gTTS(text=text, lang=lang)
    tts.save(f"/content/output_{lang}.mp3")
    return f"/content/output_{lang}.mp3"

def translate_text(text, dest_lang='hi'):
    translator = Translator()
    return translator.translate(text, dest=dest_lang).text

def speech_to_text(audio):
    recognizer = sr.Recognizer()
    with sr.AudioFile(audio) as source:
        audio_data = recognizer.record(source)
        try:
            text = recognizer.recognize_google(audio_data)
            return text
        except sr.UnknownValueError:
            return "Google Speech Recognition could not understand audio"
        except sr.RequestError as e:
            return f"Could not request results from Google Speech Recognition service; {e}"

def process(image=None, audio=None):
    if image:
        text = extract_text_from_image(image)
    elif audio:
        text = speech_to_text(audio)
    else:
        return "No input provided", None, None, None

    if text:
        english_audio = text_to_speech(text, 'en')
        hindi_text = translate_text(text, 'hi')
        hindi_audio = text_to_speech(hindi_text, 'hi')
        return text, english_audio, hindi_text, hindi_audio
    else:
        return "No text was extracted or recognized.", None, None, None

# Gradio interface
iface = gr.Interface(
    fn=process,
    inputs=[
        gr.Image(type="pil", label="Upload Image (optional)"),
        gr.Audio(source="microphone", type="filepath", label="Speak Here (optional)")
    ],
    outputs=[
        gr.Textbox(label="Extracted/Recognized Text"),
        gr.Audio(label="English Audio", type="filepath"),
        gr.Textbox(label="Translated Text (Hindi)"),
        gr.Audio(label="Hindi Audio", type="filepath")
    ],
    title="Text Extraction & Translation",
    description="Upload an image to extract text or speak to convert speech to text. Results will be translated to Hindi."
)

iface.launch()
```
